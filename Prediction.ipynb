```python
import pandas
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import numpy as np
```

```python
f = open('statistics.csv')
dataFrame_in = pandas.read_csv(f)
pandas.options.display.float_format = '{:,.0f}'.format
dataFrame_in
```

```python
dataFrame = dataFrame_in.copy()
algoList = dataFrame['Algo'].tolist()
WinningList = dataFrame['Winning'].tolist()
```

```python
labelEncoderAlgo = LabelEncoder()
labelEncoderAlgo.fit(algoList)
labelsAlgo= labelEncoderAlgo.transform(algoList)
dataFrame['Algo']=pandas.Series(labelsAlgo)
labelEncoderWin = LabelEncoder()
labelEncoderWin.fit(WinningList)
labelsWin= labelEncoderWin.transform(WinningList)
dataFrame['Winning']=pandas.Series(labelsWin)
dataFrame
```

```python
normalized_df=(dataFrame-dataFrame.min())/(dataFrame.max()-dataFrame.min())
normalized_df
```

```python
indexes = np.array(normalized_df.index.values).reshape(-1, 1)
to_predict_indexes= list(range(len(indexes), len(indexes)+5))
to_predict_indexes= np.array(to_predict_indexes).reshape(-1, 1)
score = np.array(normalized_df['Score']).reshape(-1, 1)
```

```python
regsr=make_pipeline(PolynomialFeatures(),LinearRegression())
regsr.fit(indexes, score)
```

```python
predicted_score = regsr.predict(to_predict_indexes).reshape(-1, 1)
predicted_score
```

```python
regsr.fit( np.array(normalized_df['Score']).reshape(-1, 1),  np.array(normalized_df['Time']).reshape(-1, 1))
predicted_time = regsr.predict(predicted_score)
predicted_time
```

```python
regsr = LogisticRegression()
regsr.fit( np.array(normalized_df['Score']).reshape(-1, 1) ,  np.array(normalized_df['Winning']))
predicted_winning = regsr.predict(predicted_score)
predicted_winning
```

```python
regsr.fit( np.array(normalized_df['Score']).reshape(-1, 1) ,  np.array(normalized_df['Algo']))
predicted_algo = regsr.predict(predicted_score)
predicted_algo
```

```python
new_predicted_time =[]
new_predicted_score = []
for i in range(len(predicted_time)):
    new_predicted_time.append(predicted_time[i][0])
for i in range(len(predicted_score)):
    new_predicted_score.append(predicted_score[i][0])
new_predicted_winning = []
new_predicted_algo = []
for index in range(5):
    if(list(predicted_winning)[index] == 0):
        new_predicted_winning.append(False)
    else:
        new_predicted_winning.append(True)
    if(list(predicted_algo)[index] == 0):
        new_predicted_algo.append("EXPECTIMAX")
    else:
        new_predicted_algo.append("MINIMAX")
```

```python
dict = {'Winning':new_predicted_winning, 'Time':new_predicted_time, 'Score':new_predicted_score, 'Algo':new_predicted_algo}
result = pandas.DataFrame(dict)
result
```

```python
result_df = result.copy()
result_df['Score'] = round((result_df['Score']*(dataFrame['Score'].max()-dataFrame['Score'].min()) + dataFrame['Score'].min()),0)
result_df['Time'] = (result_df['Time']*(dataFrame['Time'].max()-dataFrame['Time'].min()) + dataFrame['Time'].min())
result_df
```

```python
dataFrame_in = dataFrame_in.append(result_df, ignore_index=True)
dataFrame_in.to_csv('out.csv', index=False) 
dataFrame_in
```
